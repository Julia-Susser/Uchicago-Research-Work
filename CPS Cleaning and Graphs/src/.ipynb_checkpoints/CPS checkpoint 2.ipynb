```python
df = pd.read_stata("/Users/jsusser/Dropbox/thesis/4_CPS/clean_HPV/input/cps_00050.dta")
```

```python
df.info()
```

```python
# if sex is female, replace with 0. if sex is male, replace with 1.
df["sex"] = df["sex"].replace("female", 0)
df["sex"] = df["sex"].replace("male", 1)
df["sex"].value_counts()
```

```python
#first I replace firmsize "niu" with 0 (assumed that meant there was no firm)
df["firmsize"] = df["firmsize"].replace("niu", 0)
#next, since 0 was a category already, I could fill the nan in firm size with 0
df['firmsize'].fillna(0, inplace =True) 
for x in df["firmsize"].unique():
    if type(x) == str:
        #I stripped the values if they were a string because some of them were "100+" and took the first value
        k = x.strip("+").split()[-1]
    else:
        k = x
    if int(k)>=500:
        #if firmsize(k) is greater or equal to 100, replace with 1
        df["firmsize"] = df["firmsize"].replace(x, 1)
    else:
        #if firmsize is smaller than 500, replaced firmsize with 0
        df["firmsize"] = df["firmsize"].replace(x, 0)
```

```python
# if race is white, replace with 1. if race is not white, replace with 0.
for x in df["race"].unique():
    if x == "white":
        df["race"] = df["race"].replace(x,1)
    else:
        df["race"] = df["race"].replace(x,0)
```

```python
#This code will take the code for the occ1990 occupation and using the two digit dictionary, I found with range the value was in

#For the two digit, I returned the first value. For ex. if code was 35, it would return 23 for Management Related Occupations
def two(num):
    for x in two_digits:
        first, second = x.split("/")
        try: 
            if int(num) in range(int(first), int(second)):
                return first
        except ValueError:
            return None
        except TypeError:
            return None
#For the one_digit, I returned the occupation. For ex, 'MANAGERIAL AND PROFESSIONAL SPECIALTY OCCUPATIONS'.
#This is because I will take this title and put it into the code 1-6 later.
def one(num):
    for x,y in one_digits.items():
        first, second = x.split("/")
        try: 
            #This has a different +1 to the range because the values got sorted different, so that the start of the range is the one before it and the end is the last value of it. 
            if int(num) in range(int(first)+1, int(second)+1):
                return y
        except ValueError:
            return None
        except TypeError:
            return None

#The one digit industry function, takes the three digit industry code from below and then returns the title. For ex, 'AGRICULTURE, FORESTRY, AND FISHERIES'.
#This is because I will take this title and put it into the code 1-10 later.
def ind_one(num):
    for x,y in ind_one_digits.items():
        first, second = x.split("/")
        try: 
            if int(num) in range(int(first)+1, int(second)+1):
                return y
        except ValueError:
            return None
        except TypeError:
            return None
```

```python
#Made variable years school. 
# Ask Alex: Did NOT included Kindergarten as a year.
# Ask Alex: Count "some college but no degree" as 1 year bc most people drop out after one year. 
# Ask Alex: 5 years for doctorate. 
# Ask Alex, unknown / niu equals 0
df["years_school"] = df["educ"]
for x in df["years_school"].unique():
    if ("grade" in x) or ("none" in x) or ("high school" in x):
        #this is because when calculating the expierience variable, if the years is less than 12 then it is still 12
        df["years_school"] = df["years_school"].replace(x, 12)
        
    elif "college" in x and x != "some college but no degree":
        k = 12 + int(x.split()[0].split("+")[0])
        df["years_school"] = df["years_school"].replace(x, k)
    elif x == "some college but no degree":
        df["years_school"] = df["years_school"].replace(x, 13)
    elif x == "bachelor's degree":
        df["years_school"] = df["years_school"].replace(x, 16)
    elif (x == "master's degree") or ("associate's degree" in x):
        df["years_school"] = df["years_school"].replace(x, 18)
    elif x == "doctorate degree":
        df["years_school"] = df["years_school"].replace(x, 21)
    elif x == "professional school degree":
        df["years_school"] = df["years_school"].replace(x, 18)
    else: 
        df["years_school"] = df["years_school"].replace(x, 1)
        
        
        
```
