```python
import pandas as pd

dtafile = './Downloads/fedminwage (2).dta'
hf = pd.read_stata(dtafile)

new_values = {
    "year": [1960,1961,1962,1963,1964,1965,1966,1967], 
    "fedminwage" : [1.00,1.15,1.15,1.25,1.25,1.25,1.25,1.40]
}


df = pd.DataFrame(new_values)


# simply concatenate both dataframes 
hf = pd.concat([df, hf]).reset_index(drop = True) 



kf = pd.DataFrame({
    "year" : [int(2018), int(2019)],
    "fedminwage":[7.25,7.25]
})

hf = hf.append(kf).reset_index(drop = True) 
hf.head(60)
```

```python
for x in hf["fedminwage"]:
    hf["fedminwage"] = hf["fedminwage"].replace(x, round(x,2))
```

```python
hf.to_csv("Desktop/Fedminwage.csv")
```
